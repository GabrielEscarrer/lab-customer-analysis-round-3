```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline

file = ("files_for_lab/marketing_customer_analysis.csv")
```

```python
data = pd.read_csv(file)
```

```python
data.head()
```

```python
data.describe()
```

```python
#Show a plot of the total number of responses.
data[data.Response != "No"]
counts = data.Response.value_counts()
print(counts)
no = counts[0]
yes = counts[1]
```

```python
#Show a plot of the response rate by the sales channel.

fig, ax = plt.subplots(figsize=(14,5))
plot_1 = data[data["Response"]=="Yes"]
plot_2 = data[data["Response"]=="No"]

ax = sns.distplot(plot_1["Sales Channel"],bins=20,label="Agent")
ax = sns.distplot(plot_2["Sales Channel"],bins=20,label="Call Center")
plt.legend()

plt.show()
```

```python
#Show a plot of the response rate by the total claim amount.

fig, ax = plt.subplots(figsize=(14,5))
plot_1 = data[data["Response"]=="Yes"]
plot_2 = data[data["Response"]=="No"]

ax = sns.distplot(plot_1["Total Claim Amount"],bins=20,label="Yes")
ax = sns.distplot(plot_2["Total Claim Amount"],bins=20,label="No")
plt.legend()

plt.show()
```

```python
#Show a plot of the response rate by income.
fig, ax = plt.subplots(figsize=(14,5))
plot_1 = data[data["Response"]=="Yes"]
plot_2 = data[data["Response"]=="No"]

ax = sns.distplot(plot_1["Income"],bins=20,label="Yes")
ax = sns.distplot(plot_2["Income"],bins=20,label="No")
plt.legend()

plt.show()
```
